# Langchain

- It is a open source framework that helps in building LLM based applications
- It provides modular components & end-to-end tools that help developers build complex AI applications such as chatbots, QNA Systems, RAG, Autonomous Agents & more

### Capabilities
1. Supports all the Major LLMs.
2. Simplifies developing LLM based applications.
3. Integerations available for all major tools.
4. Open Source/ Free/ Actively Developed

### Use Cases
1. Chatbots
2. AI Knowledge Assistants
3. AI Agents
4. Workflow automations
5. Summarization/ Research Helpers

## Langchain Components:

1. Models
2. Prompts
3. Chains
4. Indexes
5. Memory
6. Agents

### 1. Models:
- In Langchain, "models" are the core interfaces through which you interact with AI models
- In Langchain, we can communicate with 2 type of models:
    1. Language Models
    2. Embedding Models

### 2. Prompts:
- In Langchain, prompts are the input provided to LLM
- Type of prompts:
    1. Dynamic & Re-usable Prompts
    2. Role-Based Prompts
    3. Few Shot Prompting

### 3. Chains:
- It helps in building pipelines
- We can create any LLM Application & trun it into chains in langchain
- Chains takes output of previous step & provide it as a input for the next step automatically
- Examples:
    1. Parallel chain pipeline
    2. Conditional chain pipeline

### 4. Indexes:
- Indexes connect your applications to external knowledge sources such as PDFs, websies or databases.
- Components:
    1. Document Loaders
    2. Text Splitter
    3. Vector Store
    4. Retrievers

### 5. Memory:
- LLM API calls are stateless i.e. Each request is independent.
- The model does NOT remember previous interactions
- You must send all required context (history, instructions, data) in every API call
- Example:
    - Who is Narendra Modi ? --> Narendra Modi is a Indian Politician
    - How Old is he ? --> The second call will not know unless you include the first message again.
- With the help of memory we can prevent these type of problems.
- Frequently used memory in Langchain:
    1. Conversation Buffer Memory - Stores a transcript of recent messages. Great for short chats but can grow large quickly.
    2. Conversation Buffer Window Memory - Only keeps the last N interactions to avoid excessive token usage.
    3. SUmmarizer based memory - Periodically summarizes older chat segments to keep a condensed memory footprint.
    4. Custom Memory - For advanced use cases, you can store specialized state in custom memory class e.g. user's preferences or key facts about them

### 6. Agents:
- It helps in building AI agents
- LLM's Capabilities --> NLU + Text Generation
- AI agents are just chatbot with superpowers i.e. Agent have reasoning capabilities with tools to perform some specific tasks.
